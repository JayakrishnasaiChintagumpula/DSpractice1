```python
# Using the LabelEncoder to convert the categorical values to numerical values
# Assigning specific numerical value to each category. This is done because, if we use dummies, it will create a new column for each category. This will increase the number of columns in the dataset.
le =  LabelEncoder()
le.fit_transform(['Married', 'Single', 'Prefer not to say'])
food1['Martial Status'] = food1['Marital Status'].map({'Married':1, 'Single':0, 'Prefer not to say':2})
```

```python
temp = foodd['Educational Qualifications'].value_counts()
temp.head()
```

```python
foodd['Educational Qualifications_new']= foodd['Educational Qualifications'].apply(lambda x: temp[x])
foodd[['Educational Qualifications','Educational Qualifications_new']].head()
```

```python
for i in range(0,len(foodd)):
    if foodd['Educational Qualifications_new'][i]>150:
        foodd['Educational Qualifications'][i]='Highly Educated'
    elif foodd['Educational Qualifications_new'][i]<=25:
        foodd['Educational Qualifications'][i]='Moderately Educated'
    else:
        foodd['Educational Qualifications_new'][i]< 5
        foodd['Educational Qualifications'][i]='Less Educated'
```
